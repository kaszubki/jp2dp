```python
airports = pd.read_csv("airports.csv", header=None)
airports.columns = ["Airport ID", "Name", "City", "Country", "IATA/FAA", "ICAO", "Latitude", "Longitude", "Altitude", "Timezone", "DST", "Tz database time zone"]
airports["Airport ID"] = [str(x) for x in airports["Airport ID"]]

routes = pd.read_csv("routes.csv", header=None)
routes.drop(routes.columns[[1, 6, 7, 8]], axis=1, inplace=True)
routes.columns = ["Airline", "Source country", "Source airport", "Destination country", "Destination airport"]

res = pd.merge(left=routes, right=airports, left_on="Source airport", right_on="Airport ID")
res.drop(res.columns[9:], axis=1, inplace=True)
res.drop(res.columns[[1, 3, 5, 7]], axis=1, inplace=True)
res.columns = ["Airline", "Source airport ID", "Destination airport ID", "Source airport", "Source country"]
res.head()
```
